### Directions can be periodic

```python
turtles_data = pd.DataFrame({'angles':[43, 45, 52, 61, 75, 88, 88, 279, 357]})
degrees = turtles_data.angles.values
```

What is the mean travelling directions?

#### 1) Naive approach

```python
def plotCircleandHist(data):
    
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(24, 8))
    
    axH = plt.subplot(121)
    axH.set_yticks([1, 2]) 
    axH.patch.set_alpha(0.0)

    axH.hist(data, bins=360)
    
    
    hist, bin_edges = np.histogram(data, bins=np.linspace(0, 360, 360))
    theta = (bin_edges[1:] + bin_edges[:-1])*0.5
    radii = hist

    width = (2*np.pi) / 180
    
    axC = plt.subplot(122, polar=True)
    bars = axC.bar(np.radians(theta), radii, width=width, color='white')

    axC.set_rticks([1, 2])  
    axC.set_rlabel_position(190) 
    axC.patch.set_alpha(0.0)
    
    
    plt.show()

plotCircleandHist(degrees)
```

```python
print ('Naive mean direction is:', np.mean(degrees))
```

#### 2) This-can't-be-right approach

The center of mass (mean) of a set of values
$ \bar{x} = \frac{x_1 + x_2 + ... x_n}{n} $

The cartesian coordinates of CoM of a set of directions

```python
Image(filename='polar.png', width=300) 
```

$\bar{C}=\frac{1}{n}\sum{\cos{\theta_i}} = \bar{R}\cos{\bar\theta}$  
and   
$\bar{S}=\frac{1}{n}\sum{\sin{\theta_i}} = \bar{R}\sin{\bar\theta}$

where $\bar{R}$ is the  _length of the mean resultant vector_ $R \in (0, 1]$


Or, use complex numbers:  $z=e^{i\theta}$  
The mean resultant vector is then:  
      $$\bar{\rho}= \frac{1}{N}\sum z_i = \bar{R} e^{i\bar{\theta}}$$


```python
from scipy.stats import circmean
print ('Mean direction is:', np.degrees(circmean(np.radians(degrees) )))
```

#### Does the naive average get less terrible for clustered data?

```python
clustered_data = pd.DataFrame({'angles':[43, 45, 52, 57, 57, 61, 63, 65, 75]})
cl_degrees = clustered_data.angles.values
plotCircleandHist(cl_degrees)
```

```python
print ('Naive average:',  np.mean(cl_degrees))
print ('Directional average:',  np.degrees(circmean(np.radians(cl_degrees))))
```
